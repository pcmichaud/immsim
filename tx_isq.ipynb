```python
import pandas as pd
import numpy as np
```

```python
pop = pd.read_excel("data/population_qc_isq.xlsx")
```

```python

pop['mx'] = pop['death']/pop['population']
pop['tx_net_interp_migration'] = pop['net_interp_migration']/pop['population']
pop['tx_net_emmigrant'] = pop['emmigrant']/pop['population']
zero_year = pop.loc[pop['age']==0].reset_index()
pop_15_45 = pop.loc[(pop['age']>=15) & (pop['age']<=45),['year','population']].groupby('year').sum().reset_index()
pop_15_45['tx_naissance'] = zero_year['population'].values/pop_15_45['population'].values
pop_15_45
```

```python
tot_immig = pop.groupby('year')['immigrant'].sum().reset_index().rename({'immigrant':'tot_immig'},axis=1)
pop = pd.merge(pop,tot_immig,on='year')
pop['prop_immig'] = pop['immigrant']/pop['tot_immig']
```

```python
pop_to_export = pop.loc[pop['year']>=2023].reset_index(drop=True)
pop_to_export.drop(['Unnamed: 0'],axis=1,inplace=True)
pop_to_export
```

```python
pop_15_45_to_export = pop_15_45.drop('population',axis=1).loc[pop_15_45['year']>=2023]
with pd.ExcelWriter('data/tx_isq.xlsx') as writer:
    pop_to_export.to_excel(writer,sheet_name='tx_by_age')
    pop_15_45_to_export.to_excel(writer,sheet_name='tx_naissance')
```
